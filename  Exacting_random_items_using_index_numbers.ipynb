## Write a program to extract the items from a list by using index numbers

```python
l = [11,22,33,42,1,133,10]

# I want items of 0,1,3,5,6 index 

index = [0,1,3,5,6]
n = []

for i in index:
    n.append(l[i])
    
print(n)
```

```python
l[0]
```

```python
l[0,1]     # we can't get more than 1 item by using indexing in list, but by using array we can get more than 1 items 
```

```python
l[[0,1]]
```

```python
# By using Numpy 
# If we want to access the multiple items at a time by using indexing then we have to pass the list of that multiple indexes for indexing
# Syntax :- a[list of indexes]

# Ex. a[[0,1,3,5,6]], array[[-1,-3,-6]],...

# we can't write it as a[0,1,3,5,6], we have to pass the list of indexes only not multiple index numbers directly
```

```python
import numpy as np

a = np.array(l)  # Coverting list to numpy array

print(a[index])
a[0]
```

```python
a[[0,1,3,5,6]]
```
